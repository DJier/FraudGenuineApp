```python
!pip install google-play-scraper

from google_play_scraper import app, reviews
import json
import time

def fetch_app_data(app_data):
    app_id = app_data.get("appId", "UNKNOWN")
    try:
        result = app(app_id, lang='en', country='in')
        user_reviews, _ = reviews(app_id, lang='en', country='in', count=10)

        return {
            "appId": app_id,
            "description": result.get("description", app_data.get("description", "N/A")),
            "title": result.get("title", app_data.get("title", "N/A")),
            "summary": result.get("summary", app_data.get("summary", "N/A")),
            "installs": result.get("installs", app_data.get("metrics", {}).get("installs", {}).get("text", "N/A")),
            "ratings": result.get("score", app_data.get("metrics", {}).get("ratings", {}).get("average", 0)),
            "permissions": result.get("permissions", app_data.get("permissions", [])),
            "reviews": [r['content'] for r in user_reviews],
            "isFraud": False,
        }

    except Exception as e:
        print(f"API failed for {app_id}. Using fallback JSON values.")
        return {
            "appId": app_id,
            "description": app_data.get("description", "N/A"),
            "title": app_data.get("title", "N/A"),
            "summary": app_data.get("summary", "N/A"),
            "installs": app_data.get("metrics", {}).get("installs", {}).get("text", "N/A"),
            "ratings": app_data.get("metrics", {}).get("ratings", {}).get("average", 0),
            "permissions": app_data.get("permissions", []),
            "reviews": [],
            "isFraud": True,
            "error": str(e)
        }

input_filename = "/content/fraud-apps.json"
with open(input_filename, 'r') as f:
    app_list = json.load(f)

enriched_apps = []

for i, app_data in enumerate(app_list):
    app_id = app_data.get("appId")
    if not app_id:
        continue

    print(f"[{i+1}/{len(app_list)}] Fetching: {app_id}")
    enriched = fetch_app_data(app_data)
    enriched_apps.append(enriched)
    time.sleep(1)

output_filename = "/content/enriched_fraud-apps.json"
with open(output_filename, "w") as f:
    json.dump(enriched_apps, f, indent=2)

total = len(app_list)
failed = [a for a in enriched_apps if "error" in a]
print(f"Total apps processed: {total}")
print(f"Apps with fetch errors: {len(failed)}")
```

```python
import re
import json
import unicodedata

def remove_emojis(text):
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"
        "\U0001F300-\U0001F5FF"
        "\U0001F680-\U0001F6FF"
        "\U0001F1E0-\U0001F1FF"
        "\U00002702-\U000027B0"
        "\U000024C2-\U0001F251"
        "]+",
        flags=re.UNICODE,
    )
    return emoji_pattern.sub(r'', text)

def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r'<.*?>', '', text)
    text = remove_emojis(text)
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8')
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return text.strip()

permission_categories = {
    'sms': ['sms', 'receive sms', 'send sms'],
    'contacts': ['contacts', 'read contacts', 'write contacts'],
    'location': ['location', 'access fine location', 'access coarse location'],
    'camera': ['camera'],
    'microphone': ['microphone'],
    'storage': ['read external storage', 'write external storage'],
    'phone': ['read phone state', 'call phone', 'read call log'],
    'calendar': ['calendar'],
    'identity': ['get accounts', 'manage accounts'],
}

def categorize_permissions(permission_list):
    categories = set()
    if not permission_list:
        return []
    for perm in permission_list:
        perm = perm.lower()
        for category, keywords in permission_categories.items():
            if any(kw in perm for kw in keywords):
                categories.add(category)
    return list(categories)

def process_file(input_filename):
    try:
        with open(input_filename, 'r', encoding='utf-8') as f:
            enriched_data = json.load(f)

        preprocessed_data = []
        for app in enriched_data:
            if not isinstance(app, dict):
                continue

            app['description_clean'] = clean_text(app.get('description', ''))
            app['summary_clean'] = clean_text(app.get('summary', ''))
            app['reviews_clean'] = [clean_text(r) for r in app.get('reviews', []) if isinstance(r, str)]
            app['permission_categories'] = categorize_permissions(app.get('permissions', []))
            preprocessed_data.append(app)

        output_filename = "preprocessed_" + input_filename.split('/')[-1]
        with open(output_filename, "w", encoding='utf-8') as f:
            json.dump(preprocessed_data, f, indent=2)

        print(f"Preprocessed {len(preprocessed_data)} apps. Saved to '{output_filename}'")

    except FileNotFoundError:
        print(f"Error: The file '{input_filename}' was not found.")
    except json.JSONDecodeError:
        print(f"Error: The file '{input_filename}' is not a valid JSON file.")
    except Exception as e:
        print(f"An unexpected error occurred while processing '{input_filename}': {e}")

files_to_process = [
    '/content/enriched_fraud-apps.json',
    '/content/enriched_genuine-apps.json'
]

for file_path in files_to_process:
    process_file(file_path)
```

```python
!pip install -q -U google-generativeai

import json
from typing import Literal
from pydantic import BaseModel
from google import genai
from google.colab import files

class AppPrediction(BaseModel):
    type: Literal["fraud", "genuine", "suspected"]
    reason: str

input_filename = "/content/preprocessed_enriched_genuine-apps.json"
with open(input_filename, 'r') as f:
    preprocessed_data = json.load(f)

client = genai.Client(api_key="AIzaSyCM9EGFkXDktrpnKnqXeXyXR93R6eLZQYQ")
model_name = "models/gemini-2.5-flash"

classified_results = []
fraud_count = 0
genuine_count = 0
suspected_count = 0

for i, app in enumerate(preprocessed_data):
    print(f"\n[{i+1}/{len(preprocessed_data)}] Processing: {app.get('title', 'UNKNOWN')}")

    try:
        installs = int(app.get("installs", "0").replace(",", "").replace("+", ""))
    except:
        installs = 0

    if app.get("isFraud", False) or installs < 20000:
        rule_hint = ""
        if app.get("isFraud", False):
            rule_hint += "This app is already flagged as fraudulent based on metadata or known developer behavior.\n"
        if installs < 20000:
            rule_hint += f"This app has very low installs ({installs}), indicating poor traction and potential fraud.\n"

        prompt = f"""
The following app is considered fraudulent.

{rule_hint}

App Name: {app.get('title')}
Description: {app.get('description_clean')}
Permissions: {app.get('permission_categories')}
Reviews: {app.get('reviews_clean')[:5]}

Based on the data above, explain why this app is likely fraudulent.
Return a structured JSON with:
- type: fraud
- reason: detailed justification
"""

        try:
            response = client.models.generate_content(
                model=model_name,
                contents=prompt,
                config={
                    "response_mime_type": "application/json",
                    "response_schema": AppPrediction,
                }
            )
            prediction = response.parsed
            if not prediction:
                raise ValueError("Model returned empty structured output.")
        except Exception as e:
            print(f"Gemini Error: {e}")
            prediction = AppPrediction(
                type="fraud",
                reason=f"Gemini error or unstructured output while explaining fraud: {str(e)}"
            )

    else:
        prompt = f"""
App Name: {app.get('title')}
Description: {app.get('description_clean')}
Permissions: {app.get('permission_categories')}
Reviews: {app.get('reviews_clean')[:5]}

Classify this app as fraud, genuine, or suspected.
Return a structured JSON with:
- type: fraud/genuine/suspected
- reason: detailed justification
"""

        try:
            response = client.models.generate_content(
                model=model_name,
                contents=prompt,
                config={
                    "response_mime_type": "application/json",
                    "response_schema": AppPrediction,
                }
            )
            prediction = response.parsed
            if not prediction:
                raise ValueError("Model returned empty structured output.")
        except Exception as e:
            print(f"Gemini Error: {e}")
            prediction = AppPrediction(
                type="suspected",
                reason=f"Gemini error or unstructured output: {str(e)}"
            )

    print(f"Result: {prediction.type} | {prediction.reason}")

    if prediction.type == "fraud":
        fraud_count += 1
    elif prediction.type == "genuine":
        genuine_count += 1
    else:
        suspected_count += 1

    app["llm_prediction"] = prediction.model_dump()
    classified_results.append(app)

output_filename = input_filename.replace("preprocessed_", "llm_classified_")
with open(output_filename, "w") as f:
    json.dump(classified_results, f, indent=2)

print("\nDONE!")
print(f"Fraud: {fraud_count}")
print(f"Suspected: {suspected_count}")
print(f"Genuine: {genuine_count}")
print(f"Total Processed: {len(classified_results)}")
```
